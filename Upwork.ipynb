```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Number of deliveries
n = 500

data = {
    "distance_km": np.random.randint(5, 300, n),
    "delivery_time_hours": np.random.normal(6, 2, n).clip(1, 15),
    "fuel_consumption_liters": np.random.normal(20, 5, n).clip(5, 50),
    "route": np.random.choice(["Route A", "Route B", "Route C", "Route D"], n),
    "vehicle_type": np.random.choice(["Van", "Truck", "Bike"], n)
}

df = pd.DataFrame(data)

# Create delay flag
df["delayed"] = df["delivery_time_hours"] > 8

df.head()
```

```python
import matplotlib.pyplot as plt

df["delayed"].value_counts().plot(kind="bar")
plt.title("On-Time vs Delayed Deliveries")
plt.xlabel("Delayed")
plt.ylabel("Number of Deliveries")
plt.savefig("on_time_vs_delayed.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.hist(df["delivery_time_hours"], bins=20)
plt.title("Delivery Time Distribution")
plt.xlabel("Delivery Time (Hours)")
plt.ylabel("Frequency")
plt.savefig("delivery_time_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.scatter(df["distance_km"], df["fuel_consumption_liters"])
plt.title("Fuel Consumption vs Distance")
plt.xlabel("Distance (km)")
plt.ylabel("Fuel Consumption (Liters)")
plt.savefig("fuel_vs_distance.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df.groupby("route")["delivery_time_hours"].mean().plot(kind="bar")
plt.title("Average Delivery Time by Route")
plt.xlabel("Route")
plt.ylabel("Average Delivery Time (Hours)")
plt.savefig("route_performance.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df.groupby("vehicle_type")["delayed"].mean().plot(kind="bar")
plt.title("Delay Rate by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Delay Rate")
plt.savefig("delay_rate_by_vehicle_type.png", dpi=300, bbox_inches="tight")
plt.show()
```
